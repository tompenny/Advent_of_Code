```python
import numpy as np
```

```python
f = open('Day5input.txt', 'r')
data = f.read()
```

# Part 1

```python
crates = data[:data.index('9')+1].split('\n')
instructions = data[data.index('9')+4:].split('\n')
crates = [[i[1], i[5], i[9], i[13], i[17], i[21], i[25], i[29], i[33]] for i in crates]
#crates = data[:data.index('3')+1].split('\n')
#instructions = data[data.index('3')+4:].split('\n')
#crates = [[i[1], i[5], i[9]] for i in crates]
crates = [list(i[::-1]) for i in np.transpose(crates)]
instructions = [[int(i[1]), int(i[3])-1, int(i[5])-1] for i in [j.split(' ') for j in instructions]]
for i in range(len(crates)):
    crates[i] = [j for j in crates[i] if j != ' '][1:]
```

```python
for i in instructions:
    for j in range(i[0]):
        crates[i[2]].append(crates[i[1]][-1])
        crates[i[1]] = crates[i[1]][:-1]
```

```python
answer = ''
for i in crates:
    print(i[-1])
```
